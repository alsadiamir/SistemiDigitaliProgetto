```python
#IMPORTS
import numpy as np
from PIL import Image
import math
```

```python
#FILTERS
blur_filter = np.array([[0.0625,0.125,0.0625],[0.125,0.250,0.125],[0.0625,0.125,0.0625]], dtype=np.float32)

#FILTER USED 
global used_filter
used_filter = blur_filter

#GLOBAL VARIABLES
global FL
FL = 8
global IL
IL = 10
global e
e = pow(2,-FL)
global n_max
n_max = pow(2,IL-1) - e
global n_min
n_min = -1*pow(2,IL-1)
```

```python
#ELABORATION
def elaboration(P):
    N = P // e
    p = (P - N*e) / e
    rand = np.random.normal(0,1,1)
    if rand <= p:
        return N*e
    else:
        return (N+1)*e
```

```python
#CONVERSION
def conversion(Z):
    if Z > n_max:
        return n_max
    elif Z < n_min:
        return n_min
    else:
        return Z
```

```python
def convert_and_compute(P):
    P = P/pow(2,FL)
    P = conversion( elaboration(P) )
    pix = int(P)
    return pix 
```

```python
#CONVOLUTION - 3x3 KERNEL
def convolution(kernel,image):
    output = image.astype(int)
    width = len(image)
    height = len(image[0])
    
    for i in range(1,width-1):
    #iterating in a column
        for j in range(1,height-1):
            #iterating in greyscale
            parz = 0.0
            ind_x = i - 1
            ind_y = j - 1
            for k in range(0,3):
                for x in range(0,3):
                    parz += image[ind_x + k][ind_y + x]*kernel[k][x]
            #print("PARZ : ",parz)
            output[i][j] = convert_and_compute(parz)
   
    return output
```
